```python
from tkinter import *
root=Tk()



def btn_click(number):
    entry.insert(END,number)

    #clear command

def btn_clear():
    entry.delete(0,END)

    #add command

def btn_add():
    first_number = entry.get()
    global f_num 
    f_num= int(first_number)
    global say
    say='add'
    entry.delete(0,END)

    #subtraction command

def btn_sub():
    first_number = entry.get()
    global f_num 
    f_num= int(first_number)
    global say
    say='sub'
    entry.delete(0,END)

    #equal command

def btn_equal():
    sec_num=entry.get()
    entry.delete(0,END)
    if say == 'add':
        entry.insert(0,f_num + int(sec_num))
    if say == 'sub':
        entry.insert(0,f_num - int(sec_num))
    

#entry
entry=Entry(root,borderwidth=5,width=30)

#1
btn_1=Button(root,padx=30,pady=10,text='1',command=lambda: btn_click(1))

#2
btn_2=Button(root,padx=30,pady=10,text='2',command=lambda: btn_click(2))

#3
btn_3=Button(root,padx=30,pady=10,text='3',command=lambda: btn_click(3))

#4
btn_4=Button(root,padx=30,pady=10,text='4',command=lambda: btn_click(4))

#5
btn_5=Button(root,padx=30,pady=10,text='5',command=lambda: btn_click(5))

#6
btn_6=Button(root,padx=30,pady=10,text='6',command=lambda: btn_click(6))

#7
btn_7=Button(root,padx=30,pady=10,text='7',command=lambda: btn_click(7))

#8
btn_8=Button(root,padx=30,pady=10,text='8',command=lambda: btn_click(8))

#9
btn_9=Button(root,padx=30,pady=10,text='9',command=lambda: btn_click(9))

#0
btn_0=Button(root,padx=30,pady=10,text='0',command=lambda: btn_click(0))

#add
add=Button(root,padx=30,pady=10,text='+',command=btn_add)

#subtract
subtract=Button(root,padx=30,pady=10,text='-',command=btn_sub)

#clear
clear=Button(root,padx=30,pady=30,text='C',command=btn_clear)

#equal
equal=Button(root,padx=30,pady=30,text='=',command=btn_equal)





#pack
entry.grid(row=0,column=1,columnspan=4)
btn_1.grid(row=1,column=0)
btn_2.grid(row=1,column=1)
btn_3.grid(row=1,column=2)
btn_4.grid(row=2,column=0)
btn_5.grid(row=2,column=1)
btn_6.grid(row=2,column=2)
btn_7.grid(row=3,column=0)
btn_8.grid(row=3,column=1)
btn_9.grid(row=3,column=2)
btn_0.grid(row=4,column=0)
add.grid(row=4,column=1)
subtract.grid(row=4,column=2)
clear.grid(row=1,column=3,rowspan=2)
equal.grid(row=3,column=3,rowspan=2)






root.mainloop()

```
